# Initial Dataset Orientation

This notebook loads the public market dataset created by the project’s download pipeline and performs a basic structural inspection. Its purpose is to confirm that the saved data can be loaded independently and to establish the dataset’s dimensions, schema, date coverage, symbol coverage, and first observations before exploratory analysis begins.

> Note: This notebook records an early dataset-orientation pass. The finalized prediction target, response variable, weights, and evaluation framework are defined in the project documentation, especially `docs/modeling_problem.md` and `docs/utility_score.md`.

```python
from pathlib import Path

import pandas as pd
```

```python
current_directory = Path.cwd()

if current_directory.name == "notebooks":
    project_root = current_directory.parent
else:
    project_root = current_directory

data_path = (
    project_root
    / "data"
    / "raw"
    / "daily_prices_2016_2025.parquet"
)

data_path
```

```python
prices = pd.read_parquet(data_path)

print("Dataset loaded successfully.")
```

```python
row_count, column_count = prices.shape

print(f"Rows: {row_count:,}")
print(f"Columns: {column_count}")
```

```python
prices.dtypes
```

```python
print(f"First date: {prices['date'].min().date()}")
print(f"Last date: {prices['date'].max().date()}")
print(f"Unique dates: {prices['date'].nunique():,}")
print(f"Unique symbols: {prices['symbol'].nunique()}")
```

```python
prices.head()
```

```python
print("Columns:")
print(prices.columns.tolist())

print("\nSymbols:")
print(sorted(prices["symbol"].unique()))
```

## Plain-English Interpretation

Each row represents the recorded market activity for one security on one trading date. It identifies the security and contains its adjusted opening, highest, lowest, and closing prices, together with its reported trading volume for that session. At this stage, the table contains only historical market observations; it does not yet include a prediction target, observation weight, or trading decision. Later processing will use information available on or before each date to construct explanatory features, while future returns will be stored separately as outcomes. A weight may also be added to represent the relative economic importance of an observation, but its precise definition has not yet been finalized.
